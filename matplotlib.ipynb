```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[1,2,3]## this is what to plot
y=[2,3,4]

plt.title('my first line graph',fontdict={'fontname':'comic sans MS','fontsize':10})
plt.xlabel('x axis')
plt.ylabel('y axis')
#plt.plot(x,y,label='next',color="yellow",linewidth=2,linestyle='--',marker='.',markersize=8,markeredgecolor='blue')
plt.plot(x,y,label='straight')
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:4],x2[:4]**2)##jitna chahe utna plot kara sakte hai



plt.xticks([0,1,2,3,4,8])#this line give tha indexing to axis
plt.yticks([1,3,4,5,7,7,8,9,16])
plt.legend()
plt.savefig('mygraph.png',dpi=550)
plt.show()
```

##barchart

```python
labels=['a','b','c']
values=[1,2,3]
b=plt.bar(labels,values)
pat=['/','o','*']
i=0
for bb in pat:
    b[i].set_hatch(bb)
    i=i+1

# b[0].set_hatch('/')
# b[1].set_hatch('o')
# b[2].set_hatch('/')
plt.figure(figsize=(6,4),)
plt.show()
```

```python
gas=pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5))
plt.plot(gas.Year,gas.USA,label='usa')
plt.plot(gas.Year,gas.Canada,label='canada')
plt.title('usa vs canada gas price')
plt.ylabel('dollar/gallon')
plt.legend()
#print(gas.year[::3]) will give year with a gap of three year
plt.show()
```

```python
for i in gas:
    if i!="Year":
        plt.plot(gas.Year,gas[i],marker='.')
plt.show()
```
